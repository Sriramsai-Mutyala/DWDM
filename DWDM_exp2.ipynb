```python
import pandas as pd
boston = pd.read_csv("/content/HousingData.csv")
```

```python
boston.head()
```

```python
df=pd.DataFrame(boston,index=None)
target_column_name="MEDV"
df[target_column_name] = boston[target_column_name]
```

```python
independent=df.columns.difference([target_column_name])
print(independent)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df[independent]=scaler.fit_transform(df[independent])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[independent],df[target_column_name])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sin=cosine_similarity(df.iloc[[0]],df.iloc[[1]])
print(cos_sin)
```
